```python
# Import Modules
import os
import pandas as pd
from pathlib import Path 
```

```python
# Read CSV File for 2023
census_data_23 = Path("Resources/census_data_2023.csv")
census_df_23 = pd.read_csv(census_data_23)

# Display the DataFrame
census_df_23.head()

# Keep only columns that contain "Median income (dollars)!!Estimate" in their name
filtered_columns_23 = [col for col in census_df_23.columns if "Median income (dollars)!!Estimate" in col or "Label (Grouping)" in col]

# Create a new DataFrame with only the filtered columns
census_df_filtered_23 = census_df_23[filtered_columns_23]

# Check the filtered DataFrame
census_df_filtered_23.head()

# Transpose DF
census_df_filtered_t_23 = census_df_filtered_23.transpose()
census_df_filtered_t_23.columns = census_df_filtered_t_23.iloc[0]
census_df_filtered_t_23.head()

# Rename Columns by taking out extra characters
census_renamed_23 = census_df_filtered_23.rename(columns=lambda col: col.replace('!!', ''))
census_renamed_23

census_renamed_23final = census_renamed_23.rename(columns=lambda col: col.replace('Median income (dollars)Estimate', ''))
census_renamed_23final

#transpose
census_t_23 = census_renamed_23final.transpose()
census_t_23.head()
```

```python
# Read CSV File for 2021
census_data_21 = Path("Resources/census_data_2021.csv")
census_df_21 = pd.read_csv(census_data_21)

# Display the DataFrame
census_df_21.head()

# Keep only columns that contain "Median income (dollars)!!Estimate" in their name
filtered_columns_21 = [col for col in census_df_21.columns if "Median income (dollars)!!Estimate" in col or "Label (Grouping)" in col]

# Create a new DataFrame with only the filtered columns
census_df_filtered_21 = census_df_21[filtered_columns_21]

# Check the filtered DataFrame
census_df_filtered_21.head()

# Transpose DF
census_df_filtered_t_21 = census_df_filtered_21.transpose()
census_df_filtered_t_21.columns = census_df_filtered_t_21.iloc[0]
census_df_filtered_t_21.head()

# Rename Columns by taking out extra characters
census_renamed_21 = census_df_filtered_21.rename(columns=lambda col: col.replace('!!', ''))
census_renamed_21

census_renamed_21final = census_renamed_21.rename(columns=lambda col: col.replace('Median income (dollars)Estimate', ''))
census_renamed_21final

#transpose
census_t_21 = census_renamed_21final.transpose()
census_t_21.head()
```

```python
census_t_21[1]
```

```python
census_21_households = census_t_21[1].reset_index()
```

```python
census_21_households.columns = census_21_households.iloc[0]
census_21_households = census_21_households[1:]
```

```python
census_21_households
```

```python
# Merge the first two DataFrames
census_21_22 = pd.merge(census_21_households, census_22_households, on="Label (Grouping)")

# Now merge the result with the third DataFrame
census_21_22_23 = pd.merge(census_21_22, census_23_households, on="Label (Grouping)")

# Check the final merged DataFrame
display(census_21_22_23.head())
```

```python
census_21_22_23_final = census_21_22_23.rename(columns=lambda col: col.replace('Median income (dollars)Estimate	', ''))

census_21_22_23_final
```

```python
census_21_22_23_final.rename(columns={"Label (Grouping)": "State name"}, inplace=True, errors="ignore")
census_21_22_23_final
```

```python
census_21_22_23_final["Average Household Income"] = 0
census_21_22_23_final
```

```python
# First, remove commas and convert the columns to numeric
columns_to_clean = ['2021 Median Household Income', '2022 Median Household Income', '2023 Median Household Income']

# Remove commas and convert to numeric
census_21_22_23_final[columns_to_clean] = census_21_22_23_final[columns_to_clean].replace({',': ''}, regex=True)

# Now convert these columns to numeric
census_21_22_23_final[columns_to_clean] = census_21_22_23_final[columns_to_clean].apply(pd.to_numeric, errors='coerce')

# Calculate the average of these columns
census_21_22_23_final['Average'] = census_21_22_23_final[columns_to_clean].mean(axis=1)

# Display the updated DataFrame
display(census_21_22_23_final.head())
```

```python
census_21_22_23_final = census_21_22_23_final.reset_index(drop=True)
census_21_22_23_final
```
